### Notes

* `pd.notna()` - hiển thị nếu các giá trị không bị thiếu
    * Các giá trị không thiếu sẽ là `True`
    * Các giá trị thiếu sẽ là `False`

### Ví Dụ

```python
df[df['salary_year_avg'].notna()]
```

`all(axis=1)`: Kiểm tra xem cả hai cột có giá trị hợp lệ (không phải NaN) hay không cho mỗi hàng. Nếu cả hai cột đều có giá trị hợp lệ, kết quả sẽ là True, nếu không, kết quả là False.
    * `axis=0`: áp dụng trên cột đó
    * `axis=` : áp dụng trên hàng đó

```python
df[df[['salary_year_avg', 'job_skills']].notna().all(axis=1)]
```

Khi dùng `df.notna()` một mình thì nó sẽ trả lại một bảng chứa cấc giá trị True, Flase. Nếu ô nào thiếu giá trị thì nó là Flase còn lại là True

```python
df.notna()
```

## Date and Time

### Datetime

#### Notes

* Cú pháp `pd.to_datetime()`: Chuyển đổi các số thành datatime 

#### Ví dụ

`errors='coerce'`: Nếu có bất kì giá trị nào không thể chuyển thành kiểu `datatime` thì nó sẽ chuyển những gaias trị đó thành `NaT (Not a Time)` để không gây lỗi

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'], errors='coerce')

df.info()
```

### Date

#### Notes
* `.dt` là một thuộc tính của pandas dùng để truy cập các phương thức và thuộc tính liên quan đến dữ liệu ngày giờ (datetime) trong một cột
* `.date` là thuộc tính trả về phần ngày (không có giờ) của đối tượng datetime

#### Ví dụ


```python
df['job_posted_date'] = df['job_posted_date'].dt.date

df.head()
```

### Groupby(): Nhóm

#### Notes:

* `groupby`: Nhóm DataFrame theo các giá trị duy nhất trong một cột cụ thể.
* Tác dụng chủ yếu để dùng các phép toán tổng hợp dễ dàng
  * `mean()`: Tính trung bình của các nhóm
  * `sum()`: Tính tổng giá trị của nhóm
  * `median()`: Tìm giá trị trung vị của mỗi nhóm
  * `min()` và `max()`: Xác định giá trị nhỏ nhất và lớn nhất trong mỗi nhóm
  * Đếm:
    * `count()`: Đếm số lượng mục không phải NA/null trong mỗi nhóm
    * `size()`: Trả về kích thước của mỗi nhóm (bao gồm cả giá trị NA/null)
  * `std()` và `var()`: Tính độ lệch chuẩn và phương sai của các nhóm
  * `first()` và `last()`: Lấy dòng đầu tiên và dòng cuối cùng của mỗi nhóm
  * `unique()`: Đếm số lượng giá trị duy nhất không phải NA/null trong mỗi nhóm

#### Ví dụ:

```python
df.groupby('job_title_short')['salary_year_avg'].mean()
```

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

### Agg()

#### Notes

* `agg()`: Có thể sử dụng nhiều hàm cùng một lúc trên nhóm hoặc sử dụng một từ điển để áp dụng các hàm cụ thể cho các cột cụ thể

#### Ví dụ

```python
df.groupby('job_title_short')['salary_year_avg'].agg(['min', 'max'])
```
